```python
import numpy as np
import matplotlib.pyplot as plt
import math
from numpy.random import normal
from scipy.stats import kstest

np.random.seed(0)
%matplotlib inline
```

```python
N = 1000
n = 50
```

```python
X = normal(0, 1, N*n).reshape(N, n)
```

```python
plt.plot(range(N), sorted([kstest(rvs=row, cdf='norm', args=(0,1), N=n)[1] for row in X]))
plt.show()
```

```python
plt.plot(range(N), sorted([kstest(rvs=row, cdf='norm', args=(row.mean(),1), N=n)[1] for row in X]))
plt.show()
```
