```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import statsmodels.api as sm

%matplotlib inline
```

```python
diabetes = datasets.load_diabetes()
```

```python
X = diabetes.data[:,[2,8]]
```

```python
X = np.concatenate((np.ones((len(X), 1)), X), axis=1)
```

```python
X_test = X[:20]
X_train = X[20:]
```

```python
Y_test = diabetes.target[:20]
Y_train = diabetes.target[20:]
```

```python
plt.scatter(X_test[:,1], Y_test, color='r')
plt.scatter(X_train[:,1], Y_train, color='b')
```

```python
plt.scatter(X_test[:,2], Y_test, color='r')
plt.scatter(X_train[:,2], Y_train, color='b')
```

```python
θ = np.linalg.inv(X_train.T.dot(X_train)).dot(X_train.T).dot(Y_train[:,np.newaxis])
```

```python
Y_pred = X_test.dot(θ).T[0]
```

```python
plt.scatter(X_test[:,1], Y_test, color='g')
plt.scatter(X_test[:,1], Y_pred, color='r')
```

```python
plt.scatter(X_test[:,2], Y_test, color='g')
plt.scatter(X_test[:,2], Y_pred, color='r')
```

```python
res = sm.OLS(Y_train, X_train).fit()
print(res.params)
print(res.conf_int(0.01))
```
